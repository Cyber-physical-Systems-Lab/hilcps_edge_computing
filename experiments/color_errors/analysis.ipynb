```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("color_error.csv")
init_timestamp = 1715102383.359453708 #copied from the csv
df["timestamp"] = df["timestamp"] - init_timestamp
df
```

```python
plt.figure(figsize=(15, 6))  # Adjust figure size if needed
plt.plot(df['timestamp'], df['camera driver'], label='Camera Driver')
plt.plot(df['timestamp'], df['controller'], label='Controller' )


plt.xlabel('Time (seconds)', fontsize=15)
plt.ylabel('Values', fontsize=15)
plt.legend(fontsize=15)

plt.yticks(range(4))
plt.grid(True, which='both', linestyle='--')  # Show vertical grids only
plt.gca().xaxis.grid(False)  # Disable horizontal grids
plt.savefig("cameravscontrollerstate")
plt.show()
```
